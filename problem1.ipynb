```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
new_df=pd.read_csv("Weather Dataset.csv")
new_df
```

```python
# # #procces the data
new_df=new_df.drop_duplicates()  #to remove the duplicates
new_df=new_df.fillna(0)  #to handle with missing data
new_df=new_df[new_df["Temperature (C)"]>-100] #to remove incorrect values
new_df=new_df[new_df["Humidity"]>=0]
new_df=new_df.sort_values("Formatted Date")


```

```python
new_df
```

```python

Temperature=np.array(new_df["Temperature (C)"])
Temperature
```

```python
Humidity=np.array(new_df["Humidity"])
Humidity
```

```python
Time=np.array(new_df["Formatted Date"])
```

```python
plt.plot(Time,Temperature,color='blue',label='blue line')
plt.title('Plot temperature over time.')
plt.ylabel('Temperature')
plt.xlabel('Time')


plt.show()
```

```python
plt.scatter(Temperature,Humidity)
plt.show
```

```python
plt.hist(Temperature)
plt.show()
```

```python
numeric_df = new_df.applymap(lambda x: pd.to_numeric(x, errors='coerce'))
numeric_df = numeric_df.dropna(axis=1, how='any')
corr=numeric_df.corr()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```
